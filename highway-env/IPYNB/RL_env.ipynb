```
import stable_baselines
stable_baselines.__version__

import gym
import highway_env

from stable_baselines import DQN

%load_ext tensorboard
import tensorflow as tf
import datetime, os
from stable_baselines.common.callbacks import CheckpointCallback
```

```
# Create and wrap the environment
env = gym.make("rl-policy-v0")
```

```
# Save a checkpoint every 1000 steps
checkpoint_callback = CheckpointCallback(save_freq=1000, save_path='.',
                                         name_prefix='rl_model')

# DQN parameters
model = DQN('MlpPolicy', env, gamma=0.8, learning_rate=5e-4, buffer_size=50000, exploration_fraction=0.1,
            exploration_final_eps=0.5, exploration_initial_eps=1.0, batch_size=32, double_q=True,
            target_network_update_freq=50, prioritized_replay=True, verbose=1, tensorboard_log="./dqn_two_lane_tensorboard/")
```

```
env.config["env"] = 12
#Train the agent for n steps
model.learn(total_timesteps=int(150000), checkpoint_callback)

#Save the trained agent
model.save('/content/drive/MyDrive/highway-env-master/dqn_RL_training_0e05')
```
